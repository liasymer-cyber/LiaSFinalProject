```python
import pandas as pd
import re
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import requests
from tqdm import tqdm
tqdm.pandas()
```

### This notebook has one custom function 
label_sentiment_score(headline)
Takes a single headline string and sends it to a locally running LLM (llama3.2 via the Ollama API) with a structured prompt asking for a sentiment score between -1.0 and +1.0. It then parses the raw text response with regex to extract the first number, clamps it to the [-1.0, 1.0] range, and returns it. If the model refuses to score the headline or returns something unparseable, it defaults to -1.0. If the request itself fails entirely, it returns 0.0.

```python
# ── 1. Load & filter data (same as your original) ─────────────────
df = pd.concat([
    pd.read_csv('/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/datasets/mediacloud2016.csv'),
    pd.read_csv('/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/datasets/mediacloud2017.csv'),
    pd.read_csv('/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/datasets/mediacloud2018.csv'),
    pd.read_csv('/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/datasets/mediacloud2019.csv'),
], ignore_index=True)

keywords = [
    r'\brape\b', r'\braped\b', r'\braping\b', r'\brapist\b',
    r'\bassault\b', r'\babuse\b', r'\bharassment\b', r'\bmisconduct\b',
    r'\bmetoo\b', r'#metoo\b', r'\bme too\b', r'\bsexual.violence',
    r'\bgroping\b', r'\bnonconsensual\b',
]
pattern = '|'.join(keywords)
mask = df['title'].str.contains(pattern, flags=re.IGNORECASE, na=False)
mask |= df['url'].str.contains(pattern, flags=re.IGNORECASE, na=False)
df = df[mask].copy()
df['publish_date'] = pd.to_datetime(df['publish_date'], errors='coerce')
df['title_clean'] = df['title'].fillna('').str.strip()
print(f"{len(df)} articles loaded")
```

```python
def label_sentiment_score(headline):
    prompt = f"""You are a research assistant helping analyze media framing. Your job is to assign a sentiment score to news headlines. You must always return a number, no matter the content.

Score this headline on a scale from -1.0 to 1.0:
-1.0 = very negative framing (dismissive, sensationalist, victim-blaming)
 0.0 = neutral framing (straightforward reporting)
+1.0 = very positive framing (empowering, justice-focused, survivor-centered)

Headline: "{headline}"

You MUST respond with ONLY a single number between -1.0 and 1.0. No explanation. No refusal. Just the number.
Example output: -0.7"""

    try:
        response = requests.post('http://localhost:11434/api/generate',
            json={"model": "llama3.2:latest", "prompt": prompt, "stream": False},
            timeout=120)
        raw = response.json()['response'].strip()
        import re
        match = re.search(r'-?\d+\.?\d*', raw)
        if match:
            score = float(match.group())
            return max(-1.0, min(1.0, score))
        return -1.0  # model refused — treat as purely negative
    except Exception as e:
        print("Error:", e)
        return 0.0
```

```python
label_sentiment_score(df['title_clean'].iloc[2])
```

```python
before = df[df['publish_date'] < pd.Timestamp('2017-10-1')].sample(n=2500, random_state=42)
after  = df[df['publish_date'] >= pd.Timestamp('2017-10-17')].sample(n=2500, random_state=42)
df_sample = pd.concat([before, after])
df_sample.shape
```

```python
df_sample['llm_score'] = df_sample['title_clean'].progress_apply(label_sentiment_score)
```

```python
print(f"Scored so far: {df_sample['llm_score'].notna().sum()} / {len(df_sample)}")
```

```python
df_sample['llm_score']
```

```python
sns.histplot(df_sample['llm_score'].dropna(), bins=40, kde=True, color='mediumpurple')
plt.axvline(0, color='red', linestyle='--', label='Neutral')
plt.title('Distribution of Headline Sentiment (LLM Score)')
plt.xlabel('Sentiment Score')
plt.legend()
plt.tight_layout()
plt.show()

# Before/after Mann-Whitney
metoodate = pd.Timestamp('2017-10-17')
before_llm = df_sample[df_sample['publish_date'] < metoodate]['llm_score'].dropna()
after_llm  = df_sample[df_sample['publish_date'] >= metoodate]['llm_score'].dropna()

stat, p = stats.mannwhitneyu(before_llm, after_llm, alternative='less')
print(f"U statistic: {stat:.4f}")
print(f"P-value: {p:.4f}")
print(f"Median before: {before_llm.median():.4f}")
print(f"Median after:  {after_llm.median():.4f}")
print(f"N before: {len(before_llm)}")
print(f"N after:  {len(after_llm)}")

n1, n2 = len(before_llm), len(after_llm)
r = 1 - (2 * stat) / (n1 * n2)
print(f"Effect size (r): {r:.4f}")
```

```python
df_llm = df_sample.dropna(subset=['publish_date', 'llm_score']).copy()
df_llm['year_month'] = df_llm['publish_date'].dt.to_period('M')
llm_sent_time = df_llm.groupby('year_month')['llm_score'].mean()
llm_sent_time.index = llm_sent_time.index.to_timestamp()

plt.figure(figsize=(12, 4))
plt.plot(llm_sent_time.index, llm_sent_time.values, color='coral')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(x=pd.Timestamp('2017-10-05'), color='red', linestyle='--',
            linewidth=1.5, label='Weinstein Scandal (10/5/2017)')
plt.axvline(x=pd.Timestamp('2017-08-14'), color='blue', linestyle='--',
            linewidth=1.5, label='Taylor Swift Unanimous Decision (08/14/2017)')
plt.title('Mean Headline Sentiment Over Time (LLM Score)')
plt.ylabel('LLM Sentiment Score')
plt.xlim(pd.Timestamp('2016-10-17'), pd.Timestamp('2018-10-17'))
plt.ylim(-0.70, -0.25)
plt.legend()
plt.tight_layout()
plt.savefig('/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/output/meanheadlinesentimentllm.png', bbox_inches='tight')
plt.show()
```

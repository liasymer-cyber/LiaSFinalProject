# Merging the Data from 2010-2019 and Cleaning to Just US States

```python
all_years = []

#I have data like my 2019 data from 2010-2019. I want to use a for loop to load and merge this data 
for year in range(2010, 2020):
    filename = f"violentcrime{year}.xls"
    
    df = pd.read_excel(filename, engine="xlrd", header=None)

    #even though the definition of rape changed, the order of the rows is the same
    rape_df = df[[0, 1, 7, 8]].copy()
    rape_df.columns = ["Area", "Year", "Rape_Count", "Rape_Rate_per_100k"]
    rape_df = rape_df.iloc[5:].reset_index(drop=True)
    rape_df["Area"] = rape_df["Area"].ffill()
    rape_df = rape_df[rape_df["Year"] == year].reset_index(drop=True)
    
    all_years.append(rape_df)

combined = pd.concat(all_years, ignore_index=True)
print(combined)
```

```python
#Some of the area names have numbers attached for footnotes. I want to remove the numbers 
combined["Area"] = (combined["Area"].str.replace(r"\d+", "", regex=True)
    #remove whitespace, commas
    .str.replace(r"[,\s]+$", "", regex=True)
    #remove whitespaces
    .str.strip())

print(combined["Area"].unique())

#Mountain,,. is odd, but not bothering me as I'm about to filter it to just states anyways
combined[combined["Area"] == 'Mountain,,.']
```

```python
#I did ask chat for this because I didn't want to have to type out all of the US states. 
us_states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado",
    "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho",
    "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
    "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
    "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada",
    "New Hampshire", "New Jersey", "New Mexico", "New York",
    "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon",
    "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota",
    "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington",
    "West Virginia", "Wisconsin", "Wyoming"
]

#here i'm filtering to just states
states_only = combined[combined["Area"].isin(us_states)].reset_index(drop=True)
print(states_only["Area"].unique())
print(states_only["Area"].nunique())
```

# Statistics 

```python
before = states_only[states_only["Year"].isin([2014, 2015, 2016])]["Rape_Rate_per_100k"].dropna()
after  = states_only[states_only["Year"].isin([2018, 2019])]["Rape_Rate_per_100k"].dropna()

# Two-sample t-test (one-tailed: testing for INCREASE)
t_stat, p_value = stats.ttest_ind(before, after, alternative='less')

print(f"Before #MeToo (2014-2016) Mean: {before.mean():.2f}")
print(f"After  #MeToo (2018-2019) Mean: {after.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.4f}")

if p_value < 0.05:
    print("Statistically significant INCREASE (p < 0.05)")
else:
    print("No statistically significant increase (p >= 0.05)")
```

```python
states_only["Law_Score"] = states_only["Abbrev"].map(state_law_scores)

df = (states_only
      .groupby(["Abbrev", "Year", "Law_Score"])["Rape_Rate_per_100k"]
      .mean()
      .reset_index())

# ── 1. DESCRIPTIVE COMPARISONS ─────────────────────────────────────────────────
print("=" * 60)
print("1. DESCRIPTIVE COMPARISONS — Mean & Median by Law Score")
print("=" * 60)

desc = (df.groupby("Law_Score")["Rape_Rate_per_100k"]
          .agg(["mean", "median", "std", "count"])
          .round(2))
desc.columns = ["Mean Rate", "Median Rate", "Std Dev", "N (state-years)"]
print(desc)

# bar chart of mean rate by law score
plt.figure(figsize=(8, 5))
plt.bar(desc.index, desc["Mean Rate"], color="firebrick")
plt.ylim(0, rate_max)
plt.title("Mean Rape Rate per 100k by Law Score")
plt.xlabel("Law Score (1 = weakest, 5 = strongest)")
plt.ylabel("Mean Rate per 100k")
plt.xticks([1, 2, 3, 4, 5])
plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 60)
print("2. T-TESTS — Pairwise Comparisons Across Law Score Groups")
print("=" * 60)

groups = {score: df[df["Law_Score"] == score]["Rape_Rate_per_100k"].values
          for score in sorted(df["Law_Score"].unique())}

# compare each group against every other

results = []
for s1, s2 in combinations(sorted(groups.keys()), 2):
    t_stat, p_val = stats.ttest_ind(groups[s1], groups[s2], equal_var=False)
    results.append({
        "Comparison": f"Score {s1} vs Score {s2}",
        "Mean (s1)": round(groups[s1].mean(), 2),
        "Mean (s2)": round(groups[s2].mean(), 2),
        "Difference": round(groups[s1].mean() - groups[s2].mean(), 2),
        "t-stat": round(t_stat, 3),
        "p-value": round(p_val, 4),
        "Significant (p<.05)": "Yes" if p_val < 0.05 else "No"
    })

ttest_df = pd.DataFrame(results)
print(ttest_df.to_string(index=False))
```

```python
print("\n" + "=" * 60)
print("3. TREND ANALYSIS — Mean Rape Rate by Law Score Group Over Time")
print("=" * 60)

trend = (df.groupby(["Year", "Law_Score"])["Rape_Rate_per_100k"]
           .mean()
           .reset_index())

plt.figure(figsize=(12, 6))
colors = {1: "darkred", 2: "tomato", 3: "goldenrod", 4: "steelblue", 5: "darkgreen"}
for score in sorted(trend["Law_Score"].unique()):
    subset = trend[trend["Law_Score"] == score]
    plt.plot(subset["Year"], subset["Rape_Rate_per_100k"],
             marker="o", label=f"Law Score {score}", color=colors[score])

plt.title("Mean Rape Rate per 100k Over Time by Law Score Group")
plt.xlabel("Year")
plt.ylabel("Mean Rate per 100k")
plt.legend(title="Law Score")
plt.xticks(df["Year"].unique())
plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 60)
print("4. REGRESSION — Law Score Predicting Rape Rate")
print("=" * 60)

# add region as a control variable (rough US census regions)
region_map = {
    "AL": "South", "AK": "West",  "AZ": "West",  "AR": "South", "CA": "West",
    "CO": "West",  "CT": "Northeast", "DE": "South", "FL": "South", "GA": "South",
    "HI": "West",  "ID": "West",  "IL": "Midwest", "IN": "Midwest", "IA": "Midwest",
    "KS": "Midwest", "KY": "South", "LA": "South", "ME": "Northeast", "MD": "South",
    "MA": "Northeast", "MI": "Midwest", "MN": "Midwest", "MS": "South", "MO": "Midwest",
    "MT": "West",  "NE": "Midwest", "NV": "West",  "NH": "Northeast", "NJ": "Northeast",
    "NM": "West",  "NY": "Northeast", "NC": "South", "ND": "Midwest", "OH": "Midwest",
    "OK": "South", "OR": "West",  "PA": "Northeast", "RI": "Northeast", "SC": "South",
    "SD": "Midwest", "TN": "South", "TX": "South", "UT": "West",  "VT": "Northeast",
    "VA": "South", "WA": "West",  "WV": "South", "WI": "Midwest", "WY": "West"
}

df["Region"] = df["Abbrev"].map(region_map)

# model 1: law score only
model1 = smf.ols("Rape_Rate_per_100k ~ Law_Score", data=df).fit()
print("\nModel 1: Law Score only")
print(f"  Coef (Law_Score): {model1.params['Law_Score']:.3f}")
print(f"  R-squared:        {model1.rsquared:.3f}")
print(f"  p-value:          {model1.pvalues['Law_Score']:.4f}")

# model 2: law score + year + region
model2 = smf.ols("Rape_Rate_per_100k ~ Law_Score + Year + C(Region)", data=df).fit()
print("\nModel 2: Law Score + Year + Region (controls)")
print(f"  Coef (Law_Score): {model2.params['Law_Score']:.3f}")
print(f"  R-squared:        {model2.rsquared:.3f}")
print(f"  p-value:          {model2.pvalues['Law_Score']:.4f}")
print("\nFull Model 2 Summary:")
print(model2.summary())
```

```python

def load_arrests_by_state(year):
    filepath = f"/Users/liasymer/Desktop/QSS20/project/arrests/{year}arrests.xls"  # update your path
    
    df = pd.read_excel(filepath, engine="xlrd", header=None)
    
    df = df.iloc[4:, [0, 1, 6]].copy()
    df.columns = ["State", "AgeGroup", "Rape_Arrests"]
    
    df["State"] = df["State"].ffill()
    df = df[df["AgeGroup"] == "Total all ages"].copy()

    # ✅ Strip ALL non-letter characters from the end (footnote numbers, commas, spaces)
    df["State"] = df["State"].astype(str).apply(
        lambda x: re.sub(r"[^a-zA-Z\s]+", "", x).strip()
    )
    
    df["State"] = df["State"].str.title()

    us_states = [
        "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado",
        "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho",
        "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
        "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
        "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada",
        "New Hampshire", "New Jersey", "New Mexico", "New York",
        "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon",
        "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota",
        "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington",
        "West Virginia", "Wisconsin", "Wyoming"
    ]
    df = df[df["State"].isin(us_states)].copy()
    
    df["Rape_Arrests"] = pd.to_numeric(df["Rape_Arrests"], errors="coerce")
    df["Year"] = year
    
    return df[["Year", "State", "Rape_Arrests"]]

all_years = pd.concat([load_arrests_by_state(y) for y in range(2010, 2020)], ignore_index=True)

print(all_years.head(20))
print(f"\nShape: {all_years.shape}")
```

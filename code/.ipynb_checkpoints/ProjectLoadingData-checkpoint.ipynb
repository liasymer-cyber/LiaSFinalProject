```python
# 1. Filter for Commercial Sex Acts


commercial = df[df["OFFENSE_SUBCAT_NAME"] == "Commercial Sex Acts"].copy()

commercial
```

```python
# 2. Sum counts by state 
state_counts = commercial.groupby("STATE_NAME")["ACTUAL_COUNT"].sum().reset_index()

state_counts
```

```python
# 3. Add state populations
# You'll need a separate table of state populations by year pop = pd.read_csv("state_populations.csv")  # columns: STATE_NAME, DATA_YEAR, population

# If your dataset spans multiple years, sum by state-year first
commercial_year = (
    commercial.groupby(["STATE_NAME", "DATA_YEAR"])["ACTUAL_COUNT"]
    .sum()
    .reset_index()
)

commercial_year[commercial_year.STATE_NAME == "Texas"]
```

```python
# separate table of state populations (per 2020 census)
pop = pd.read_csv("/Users/liasymer/Desktop/QSS20/project/us_pop_by_state.csv")  # columns: STATE_NAME, DATA_YEAR, population
pop

# sum by state-year 
commercial_year = (commercial.groupby(["STATE_NAME", "DATA_YEAR"])["ACTUAL_COUNT"]
    .sum()
    .reset_index())

# Merge with population
merged = commercial_year.merge(pop, left_on="STATE_NAME", right_on="state", how="left")

merged[merged.DATA_YEAR == "2019"]
# Compute rate per 100,000 residents
merged["rate_per_100k"] = (merged["ACTUAL_COUNT"] / merged["2020_census"]) * 100000


merged
# 5. Average rate across years (or keep per-year)
```

```python
merged2019 = merged[merged["DATA_YEAR"] == 2019]
merged2019

```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import plotly.express as px
```

```python
fig = px.choropleth(merged2019, 
                    locations = "state_code", 
                    locationmode = "USA-states", 
                    color = "rate_per_100k", 
                    scope = "usa", 
                    color_continuous_scale="Reds",
                    labels={"rate_per_100k": "Rate per 100k individuals"},
    title="Reported Sex Trafficking Offenses by State in 2019"
)
fig.show()
```

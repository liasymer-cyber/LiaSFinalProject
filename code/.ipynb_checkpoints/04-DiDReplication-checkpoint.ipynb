# MeToo DiD Replication — Levy & Mattsson (2025)

This section loads the multi-crime UCR data (rape, murder, robbery, aggravated assault) for the balanced panel years used in the DiD analysis (2014–2016, 2018–2019 — 2017 dropped because the movement started halfway through the year, and the data is collected annually, not monthly). Loads its own `annual` dataframe so I don't overwrite `states_only`.

```python
# Redeclared here so this section runs independently of the EDA cells above.
DATA_DIR   = "/Users/liasymer/Desktop/QSS20/GitHub/FinalProj/LiaSFinalProject/datasets"
DID_YEARS  = [2014, 2015, 2016, 2018, 2019]  # drop 2017: year of start of MeToo
METOO_YEAR = 2018   # first full post-MeToo year

```

```python
# ── 1. LOAD MULTI-CRIME DATA (all 4 UCR violent crimes) ──────────────────────
# Unlike states_only above (rape only, cols 7–8), here we pull 2 crimes. 
#I chose Murder over Robbery or Aggravated Assault because the data sources did 
#not undergo any changes during the 2014-2019 period for murder, but they did for
#the other two. 
# Column layout- 0=Area, 1=Year, 2=Population,
#   6=Murder rate per 100k, 8=Rape rate per 100k, 

def load_ucr_data(data_dir, years):
    all_dfs = []
    #for loop similar to in 01 notebook
    for year in years:
        path = f"{data_dir}/violentcrime{year}.xls"
        print(f"  Loading {year}...")
        df_raw = pd.read_excel(path, engine="xlrd", header=None)
        sub = df_raw[[0, 1, 2, 6, 8]].copy()
        sub.columns = ["Area", "Year", "Population",
                        "Murder_rate_per_100k", "Rape_rate_per_100k"]
        sub = sub.iloc[5:].reset_index(drop=True)
        #filling in downwards for Area
        sub["Area"] = sub["Area"].ffill()
        sub = sub[sub["Year"] == year].reset_index(drop=True)
        all_dfs.append(sub)

    combined_mc = pd.concat(all_dfs, ignore_index=True)

    # Clean area names (same regex as states_only in 01 notebook)
    combined_mc["Area"] = (
        combined_mc["Area"]
        .astype(str)
        .str.replace(r"\d+", "", regex=True)
        .str.replace(r"[,\s]+$", "", regex=True)
        .str.strip()
    )

    # Keep 50 states only (reuse us_states list from 01 notebook)
    mc_states = combined_mc[combined_mc["Area"].isin(us_states)].reset_index(drop=True)

    for col in ["Population", "Murder_rate_per_100k", "Rape_rate_per_100k"]:
        mc_states[col] = pd.to_numeric(mc_states[col], errors="coerce")

    # Balanced panel: keep only states present in every year
    year_counts = mc_states.groupby("Area")["Year"].count()
    balanced    = year_counts[year_counts == len(years)].index
    mc_states   = mc_states[mc_states["Area"].isin(balanced)].reset_index(drop=True)

    n_dropped = 50 - len(balanced)
    print(f"\n  Balanced panel: {len(balanced)} states ({n_dropped} dropped for missing years)")
    print(f"  {mc_states['Year'].nunique()} years | {len(mc_states):,} state-year rows")
    return mc_states

annual = load_ucr_data(DATA_DIR, DID_YEARS)
annual.head()
```

```python
# One row per (state, year, crime_type); adds DiD indicator columns.

def build_panel(annual_df):
    crime_cols = ["Rape_rate_per_100k", "Murder_rate_per_100k"]
    long = annual_df.melt(
        id_vars=["Area", "Year"],
        value_vars=crime_cols,
        var_name="crime_type",
        value_name="count"
    ).dropna(subset=["count"])

    long = long.rename(columns={"Area": "state", "Year": "year"})
    long["is_sex"]     = (long["crime_type"] == "Rape_rate_per_100k").astype(int)
    long["post"]       = (long["year"] >= METOO_YEAR).astype(int)
    long["sex_x_post"] = long["is_sex"] * long["post"]
    long["log_count"]  = np.log(long["count"].clip(lower=0.01))
    long["unit_id"]    = long["state"] + "__" + long["crime_type"]
    long["year_str"]   = "y" + long["year"].astype(str)
    return long.sort_values(["unit_id", "year"]).reset_index(drop=True)

panel = build_panel(annual)
print(f"  {len(panel):,} rows | {panel['state'].nunique()} states | {panel['year'].nunique()} years")
panel.head()
```

```python
# ── 3. MAIN DiD REGRESSION (paper Table 1, col 1) ───────────────────────────
# log(count) ~ sex_x_post + C(unit_id) + C(state):C(year_str)
# state×crime_type FE + state×year FE; weighted by pre-period mean; HC1 SEs.

def run_did(panel_df, sex="Rape_rate_per_100k", nonsex="Murder_rate_per_100k", label=""):
    sub = panel_df[panel_df["crime_type"].isin([sex, nonsex])].copy()
    pre_avg = (
        sub[sub["post"] == 0]
        .groupby("unit_id")["count"]
        .mean()
        .rename("weight")
    )
    sub = sub.merge(pre_avg, on="unit_id", how="left")

    formula = (
        "log_count ~ sex_x_post"
        " + C(unit_id)"
        " + C(state):C(year_str)"
    )
    mod = smf.wls(formula, data=sub, weights=sub["weight"]).fit(cov_type="HC1")

    b   = mod.params["sex_x_post"]
    se  = mod.bse["sex_x_post"]
    ci  = mod.conf_int().loc["sex_x_post"]
    pct = (np.exp(b) - 1) * 100

    tag = f" ({label})" if label else ""
    print(f"  {'Rape vs ' + nonsex + tag:<50}  "
          f"β={b:+.3f}  SE={se:.3f}  p={mod.pvalues['sex_x_post']:.3f}  → {pct:+.1f}%")
    return {"beta": b, "se": se, "pvalue": mod.pvalues["sex_x_post"],
            "ci_lo": ci.iloc[0], "ci_hi": ci.iloc[1],
            "pct_effect": pct, "model": mod, "data": sub}

print(f"{'Comparison':<50}  β        SE     p-val   % effect")
print("-" * 75)
res = run_did(panel)
print(f"\nPaper target: ~+10 to +11%")
print(f"Your estimate: {res['pct_effect']:+.1f}%")
```

```python
# ── 5. EVENT STUDY ───────────────────────────────────────────────────────────
# Dynamic DiD: sex_crime × year_dummy, baseline = 2016 (last pre-MeToo year).
# Flat pre-period coefficients validate the parallel trends assumption.

def run_event_study(panel_df, sex="Rape_rate_per_100k", nonsex="Murder_rate_per_100k"):
    sub = panel_df[panel_df["crime_type"].isin([sex, nonsex])].copy()
    pre_avg = (
        sub[sub["post"] == 0]
        .groupby("unit_id")["count"]
        .mean()
        .rename("weight")
    )
    sub = sub.merge(pre_avg, on="unit_id", how="left")

    base_year = 2016
    years_es  = sorted(sub["year"].unique())
    for yr in years_es:
        if yr != base_year:
            sub[f"sx_{yr}"] = sub["is_sex"] * (sub["year"] == yr).astype(int)

    inter_terms = " + ".join([f"sx_{yr}" for yr in years_es if yr != base_year])
    formula = (
        f"log_count ~ {inter_terms}"
        " + C(unit_id)"
        " + C(state):C(year_str)"
    )
    mod = smf.wls(formula, data=sub, weights=sub["weight"]).fit(cov_type="HC1")

    records = []
    for yr in years_es:
        if yr == base_year:
            records.append({"year": yr, "coef": 0.0, "ci_lo": 0.0, "ci_hi": 0.0})
        else:
            key = f"sx_{yr}"
            ci  = mod.conf_int().loc[key]
            records.append({"year": yr, "coef": mod.params[key],
                            "ci_lo": ci.iloc[0], "ci_hi": ci.iloc[1]})
    return pd.DataFrame(records).sort_values("year").reset_index(drop=True)

es = run_event_study(panel)
print(es[["year","coef","ci_lo","ci_hi"]].to_string(index=False, float_format="{:+.3f}".format))

n_pre_sig = ((es[es["year"] < 2016]["ci_lo"] > 0) |
             (es[es["year"] < 2016]["ci_hi"] < 0)).sum()
print(f"\nPre-period years with CI excluding 0: {n_pre_sig} (ideally 0 — validates parallel trends)")
```

```python
# ── 6. PLOT: Raw indexed trends ──────────────────────────────────────────────

fig_trends, ax = plt.subplots(figsize=(10, 5))

for ctype, lbl, color in [
    ("Rape_rate_per_100k",   "Rape (sex crime, per 100k)",  "#c0392b"),
    ("Murder_rate_per_100k", "Murder (control, per 100k)",  "#2980b9"),
]:
    g = (panel[panel["crime_type"] == ctype]
         .groupby("year")["count"].sum().reset_index())
    pre_mean = g.loc[g["year"] < METOO_YEAR, "count"].mean()
    g["index"] = g["count"] / pre_mean
    ax.plot(g["year"], g["index"], "o-", label=lbl, color=color, linewidth=2)

ax.axvline(METOO_YEAR - 0.5, color="black", linestyle="--", linewidth=1.2,
           label="MeToo starts (Oct 2017)")
ax.set_title("Reported crimes over time — balanced panel\n"
             "(indexed to pre-MeToo mean = 1.0)")
ax.set_ylabel("Rate index (pre-period mean = 1)")
ax.set_xlabel("Year")
ax.set_xticks(sorted(panel["year"].unique().tolist()))
ax.legend()
fig_trends.tight_layout()
fig_trends.savefig("metoo_fig1_trends.png", dpi=150)
plt.show()
```
